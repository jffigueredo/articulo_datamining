```python
import pandas as pd

# Cargar la base de datos (reemplaza 'nombres.csv' con tu archivo)
df = pd.read_csv('nombres.csv')

# Crear listas de nombres masculinos y femeninos (ejemplos)
nombres_masculinos = ['Juan', 'Carlos', 'Pedro', 'Luis']
nombres_femeninos = ['Maria', 'Ana', 'Laura', 'Sofia']

# Función para asignar género
def asignar_genero(nombre):
    if nombre in nombres_masculinos:
        return 'Masculino'
    elif nombre in nombres_femeninos:
        return 'Femenino'
    else:
        return 'Desconocido'

# Aplicar la función a la columna de nombres
df['Genero'] = df['Nombre'].apply(asignar_genero)

# Imprimir el DataFrame con la columna de género
print(df)

#Guardar el resultado en un nuevo archivo csv.
df.to_csv('nombres_con_genero.csv', index=False)
```
